# Imports


````python
# Ben√∂tigte Imports
import tensorflow as tf
import pandas as pd
from keras.models import Sequential
from keras.layers import InputLayer, Dense, Activation
from sklearn.preprocessing import MinMaxScaler

# used in Own Predictions
import numpy as np

````

# Prepreocessing

## Read Data

````python
# Einlesen der Daten
data = pd.read_csv('/content/drive/MyDrive/dev/Titanic.csv', encoding='utf-8')

````

## Convert String data to int Values

````python
d = {0.0: 0, '1': 1}
data['Survived'] = data['Survived'].map(d).fillna(data['Survived'])
data.head()
````

````python
d = {'No': 0, 'Yes': 1}
data['Survived'] = data['Survived'].map(d).fillna(data['Survived'])
data.head()
````

````python
d = {'Male': 1, 'Female': 0}
data['Sex'] = data['Sex'].map(d).fillna(data['Sex'])
data.head()
````

````python
d = {'1st': 1, '2nd': 2, '3rd': 2, 'Crew': 0}
data['Class'] = data['Class'].map(d).fillna(data['Class'])
data.head()
````

````python
d = {'Child': 0, 'Adult': 1}
data['Age'] = data['Age'].map(d).fillna(data['Age'])
data.head()
````

````python
data
````

## Splitting the dataset in X and Y 


````python
# Zielvariable (target) und Features (input) ausw√§hlen
target = data['Survived']
input = data.drop('Survived', axis=1, )

# Daten skalieren
scaler = MinMaxScaler(feature_range=(0, 1))
scaler.fit(input)
input = scaler.transform(input)
target = target.values
````

# ML Modell

## Initialazing Modell


````python
# Modell erstellen
model = Sequential()

# Input Layer
model.add(InputLayer(name = "InputLayer", input_shape=(5,)))

# Hidden Layer
model.add(Dense(units=10, name = "HiddenLayer1", activation="relu"))
model.add(Dense(units=5, name = "HiddenLayer2", activation="relu"))

# Output Layer
model.add(Dense(units=1, name = "OutputLayer", activation="sigmoid"))
````

## Compilling Modell 


````python
# Modell konfigurieren
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

# Modell trainieren
model.fit(x = input, y = target, epochs = 500)
````

# Own Predictions

````python
hypMann = [0, 1, 1, 1, 1]
hypFrau = [0, 0, 1, 0, 1]
model.predict(np.array([hypMann, hypFrau]))
````

````python
hypMann = [2, 1, 1, 1, 1]
hypFrau = [2, 0, 1, 0, 1]
model.predict(np.array([hypMann, hypFrau]))
````
